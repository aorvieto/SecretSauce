### Importing Data

```python
file = 'Signum_160M_SP_SL512_2Btok_BS256'
res1 = pd.read_pickle('../data/'+file+'.pkl')
res1
```

```python
lrs = [0.0000625, 0.000125, 0.00025, 0.0005, 0.001, 0.002]
clipz = [1.0]
beta1s_n = [0.9875, 0.975, 0.95, 0.9, 0.8]

fig, axes = plt.subplots(1, 5, figsize=(11, 3.0), sharey=True)

for j, beta1 in enumerate(beta1s_n):
    ax = axes[j]
    isclip=['true','false']
    for i, clipp in enumerate(clipz):
        linestyle = '-' if i == 0 else '--'
        subset = res1[(res1['clip'] == clipp) & (res1['beta1'] == beta1)]
        lrs_plt = np.array(subset['lr'])
        data = np.array(subset['ppl_test'])
        ax.plot(lrs_plt, data, linestyle=linestyle, marker='o', color=colors_b[2], label=f'clip = {isclip[i]}')

    ax.set_xscale('log')
    ax.set_ylim([24.2, 28.5])
    ax.set_title(r'Signum+wd, $\beta$ = ' + str(beta1), fontsize=11)
    ax.set_xlabel('learning rate')
    ax.grid(True)

axes[0].set_ylabel('final test ppl')

plt.tight_layout()
#plt.savefig('160M_signEMA_512.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
